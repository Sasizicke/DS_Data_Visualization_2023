```python
who = pd.read_csv('who_aap_2021_v9_11august2022.csv', sep=";")
who = who.rename(columns={"Measurement Year" : "Year", "WHO Country Name" : "Country", 'ISO3' : 'Code' })
who['PM10'] = who['PM10 (μg/m3)'].apply(lambda x : float(x))
```

```python
who = who.groupby((['Country', 'Year', 'Code'])).agg({'PM10':np.median}).reset_index()
#who = who.drop(columns=['Status','Version of the database','Number and type of monitoring stations','Reference', 'NO2 temporal coverage (%)','WHO Region', 'ISO3', 'City or Locality', 'PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)', 'PM25 temporal coverage (%)', 'PM10 temporal coverage (%)' ])
```

```python
who.to_csv('who_aap_2021_v9_11august2022_clean.csv')
```
